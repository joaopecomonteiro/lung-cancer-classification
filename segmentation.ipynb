```python
#Caminho para a pasta com as imagens

dicom_dir = Path(r"C:\Users\joaom\Desktop\manifest-1600709154662\LIDC-IDRI")
```

```python
#Lista com todos os caminhos para a pasta de cada paciente

all_patients = list(dicom_dir.glob("LID*"))
all_patients[0:5]
```

```python
"""
Criar uma lista com todos os pacientes, cada paciente é um objeto da classe Patient do ficheiro utils.py
O patient_scan é um objeto da class scan da biblioteca pylidc, logo é recomendado que se consulte a documentação 
desta biblioteca para se perceber melhor como funcionam algumas das funções usadas (https://pylidc.github.io/).
"""

patient_list = []

for path in all_patients:
    
    patient_id = str(path)[-14:]
    print(f"Patient: {patient_id}")
    patient_scan = pl.query(pl.Scan).filter(pl.Scan.patient_id == patient_id).first() #Isto é o scan
    patient_nods = patient_scan.cluster_annotations()
    patient = utils.Patient(patient_id, path, patient_scan, patient_nods)
    patient_list.append(patient)
```

```python
"""
Criar uma pasta para cada paciente que contém uma pasta para cada nódulo.
A pasta de um nódulo contém uma pasta para cada anotação.
A pasta de uma anotação contém a imagem do nódulo, o ficheiro image.nrrd, a imagem da segmentação, o ficheiro mask.nrrd
e a malignancy do nódulo de acordo com aquela anotação, o ficheiro malignancy.txt.
"""
for patient in patient_list:
    
    patient_folder = f"C:/Users/joaom/Desktop/trabalho_liacd/images_and_masks/{patient.idx}"
    print(f"PATIENT {patient.idx}")
    
    for i, nod in enumerate(patient.nods):
        
        print(f"Nodule {i+1}")
        nodule_folder = f"{patient_folder}/nodule{i+1}"  
        if not os.path.exists(nodule_folder):
            os.makedirs(nodule_folder)
            
        for i, ann in enumerate(nod):
            
            annotation_folder = f"{nodule_folder}/annotation{i+1}"
            if not os.path.exists(annotation_folder):
                os.makedirs(annotation_folder)
            
            utils.get_nrrd_images(ann, i+1, annotation_folder)

            print("---------------------------")
```
